```python
# imports
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
import requests
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2:1b'
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("OPENAI_API_KEY not found")
elif not api_key.startswith('sk-proj-'):
    print("api key found but youre not using the correct key")
elif api_key.strip() != api_key:
    print("please remove spaces from your api key")
else:
    print("api key found and is valid")

openai = OpenAI()
```

```python
# Get gpt-4o-mini to answer, with streaming
system_prompt = """
You are a helpful assistant that can answer questions about the code. You explain the code in a way that is easy to understand and you answer short and concise.
"""
```

```python
# Get Llama 3.2:1b to answer
requests.get("http://localhost:11434").content
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
def stream_response(m):
    question = input("Please enter your question: ")

    if m == MODEL_GPT:
        stream = openai.chat.completions.create(
            model=m,
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": "Give me the name of a pokemon after you answer the following question: " + question}
            ],
            stream=True
        )    
    elif m == MODEL_LLAMA:
        stream = ollama.chat.completions.create(
            model=m,
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": question}
            ],
            stream=True
        )
        
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
#gpt
stream_response(MODEL_GPT)
```

```python
#llama
stream_response(MODEL_LLAMA)
```
